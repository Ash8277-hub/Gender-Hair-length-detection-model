```python
import os
import numpy as np
import pandas as pd
import cv2
from sklearn.cluster import KMeans
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.models import Model
```

```python
# Load VGG16 and remove the top layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
model = Model(inputs=base_model.input, outputs=base_model.layers[-1].output)
```

```python
# Directory of UTKFace images
data_dir = 'C:/Ashrit/Dataset_1/hairdata'
image_paths = [os.path.join(data_dir, fname) for fname in os.listdir(data_dir) if fname.endswith('.jpg')]
```

```python
# Extract features from images
features = []
for path in image_paths:
    img = load_img(path, target_size=(128, 128))
    img_array = img_to_array(img)
    img_array = preprocess_input(np.expand_dims(img_array, axis=0))
    feature = model.predict(img_array)
    features.append(feature.flatten())

features = np.array(features)
```

```python
# Apply K-Means clustering to separate hair lengths
kmeans = KMeans(n_clusters=2, random_state=0).fit(features)
labels = kmeans.labels_
```

```python
# Map cluster labels to "long" or "short" based on inspection
mapped_labels = ["long" if label == 1 else "short" for label in labels]
```

```python
# Save annotated data to CSV
annotations = pd.DataFrame({"filename": os.listdir(data_dir), "hair_length": mapped_labels})
annotations.to_csv('C:/Users/Ashrit/Desktop/ML Prjct 2/Hair_length_model/hair_length_.csv', index=False)
```
